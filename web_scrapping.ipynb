```python
from bs4 import BeautifulSoup as bs
```

```python
import requests
```

```python
link="https://www.amazon.in/Nike-Air-Max-90-QS/product-reviews/B07VXY2GFB/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"
```

```python
page=requests.get(link)
page
```

```python
page.content
```

```python
raw_data=bs(page.content,"html.parser")
```

```python
raw_data
```

```python
names=raw_data.find_all("span",class_="a-profile-name")
```

```python
names
```

```python
cust_name=[]
for i in range(len(names)):
    cust_name.append(names[i].get_text())
```

```python
cust_name
```

```python
reviews=raw_data.find_all("span",class_="a-size-base review-text review-text-content")
```

```python
reviews
```

```python
len(reviews)
```

```python
cust_reviews=[]
for i in range(len(reviews)):
    cust_reviews.append(reviews[i].get_text())
```

```python
cust_reviews
```

```python
customer_reviews=[reviews.strip("\n") for reviews in cust_reviews]
```

```python
customer_reviews
```

```python
import pandas as pd
```

```python
df=pd.DataFrame()
```

```python
df["Name"]=cust_name
df["Reviews"]=customer_reviews
```

```python
df
```
